```python
import pandas as pd
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
from tqdm import tqdm

tqdm.pandas()
stemmer = PorterStemmer()
lemmatizer=WordNetLemmatizer()
```

## Directional Words

```python
FILE_DIR = "./custom_lexicons/sentidd/"
DIRECTIONAL_WORDS_RAW_FILEPATH = FILE_DIR+'directional_words_raw.csv'
DIRECTIONAL_WORDS_FILEPATH = FILE_DIR+'directional_words.csv'

directional_words_df = pd.read_csv(DIRECTIONAL_WORDS_RAW_FILEPATH)
directional_words_df['stemmed'] = directional_words_df['token'].progress_apply(lambda x: stemmer.stem(x))
directional_words_df
```

```python
directional_words_df.to_csv(DIRECTIONAL_WORDS_FILEPATH, index=False)
```

## Loughran-McDonald Dictionary

```python
LM_RAW_FILEPATH = "Loughran-McDonald_MasterDictionary_1993-2021.csv"
LM_FILEPATH = "LM_Word_List.csv"
lm_df = pd.read_csv(FILE_DIR+LM_RAW_FILEPATH)
lm_df
```

```python
lexicon_df = lm_df.query('Negative != 0 | Positive != 0')
lexicon_df = lexicon_df[["Word", "Negative", "Positive"]]
lexicon_df
```

```python
len(lm_df[lm_df['Negative']!=0]) + len(lm_df[lm_df['Positive']!=0])
```

```python
lexicon_df["word"] = lexicon_df["Word"].str.lower()
lexicon_df["label"] = ["positive" if positive!=0 else "negative" for positive in lexicon_df["Positive"]]
lexicon_df = lexicon_df[["word","label"]]
lexicon_df
```

```python
lexicon_df.to_csv(FILE_DIR+LM_FILEPATH, index=False)
```
